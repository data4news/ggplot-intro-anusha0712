### Box and Whisker Plot
http://www.sthda.com/english/wiki/ggplot2-box-plot-quick-start-guide-r-software-and-data-visualization



````python
%%R -w 1000 -h 200

ggplot(df) + 
    aes(x=pct_below_poverty, y=1) +
    geom_boxplot(alpha=.2,outlier.size=10)
````

Box and whisker plots aren't super easy to read. We don't always use them in reader-facing visualizations, but you'll find them a lot in scientific papers. Box and whisker plots help draw attention to outliers in relation to the median and interquartile range of the data.

````python
%%R -w 1000 -h 200

# modify this code
# Anusha - so I checked stackoverflow and chatgpt and this logic made the most sense to me (so I will remember)
# %>% pipe symbol - passes input on function into another. Can just use df as parameter too.
# 1.5 * IQR is like the outliers beyond the 3rd quartile 
# quantile(x, probability between 0 and 1) so 0.25 is first quartile, 0.75 is third
# so it calculates outliers before the 1st quartile and after the third. Here we didn't need those before 0.25
ggplot(df) + 
    aes(x=pct_below_poverty, y=1) +
    geom_boxplot(alpha=.2,outlier.size=10) +
    geom_text(
    data = filter(df, pct_below_poverty > quantile(pct_below_poverty, 0.75) + 1.5 * IQR(pct_below_poverty) |
                         pct_below_poverty < quantile(pct_below_poverty, 0.25) - 1.5 * IQR(pct_below_poverty)),
    aes(label = pct_below_poverty), # binds this aesthetic to the geom_text
    vjust = -2, 
    hjust = 0.4,
    size = 4, 
    color = "red"
  )
````

````python
%%R -w 1000 -h 200

# modify this code

ggplot(df) + 
    aes(x=pct_below_poverty, y=1) +
    geom_boxplot(alpha=.2,outlier.size=10) +
    geom_text(
    data = filter(df, pct_below_poverty > quantile(pct_below_poverty, 0.75) + 1.5 * IQR(pct_below_poverty) |
                         pct_below_poverty < quantile(pct_below_poverty, 0.25) - 1.5 * IQR(pct_below_poverty)),
    aes(label = borough), # binds this aesthetic to the geom_text
    vjust = -2, 
    hjust = 0.4,
    size = 4, 
    color = "red"
  )
````
